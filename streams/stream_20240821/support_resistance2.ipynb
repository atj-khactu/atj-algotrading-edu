```python
import MetaTrader5 as mt5
from IPython.display import display
from datetime import datetime
import plotly.express as px

from atj_trading.backtester import get_ohlc_history, create_price_fig
from config import mt5_credentials
```

```python
mt5.initialize()
mt5.login(mt5_credentials['login'], mt5_credentials['password'], mt5_credentials['server'])
```

```python
symbol = 'EURUSD'
timeframe = mt5.TIMEFRAME_D1

start_dt = datetime(2023, 1 ,1)
end_dt = datetime.now()

ohlc = get_ohlc_history(symbol, timeframe, start_dt, end_dt)
display(ohlc)

create_price_fig(ohlc.tail(400))
```

```python
ohlc2 = ohlc.tail(400).copy()

period = 10

ohlc2['support'] = ohlc2['low'].shift(1).rolling(period).min()
ohlc2['resistance'] = ohlc2['high'].shift(1).rolling(period).max()

create_price_fig(ohlc2, indicators=['support', 'resistance'])
```
